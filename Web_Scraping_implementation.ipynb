**obtain HTML using Beautiful Soup**

```python
import requests
from bs4 import BeautifulSoup
url="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
r=requests.get(url)
print(r)
```

```python
print(r.text)
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

**TAGS IN HTML**

```python
print(soup.div)
```

```python
print(soup.div.p)
```

```python
print(soup.div.a)
```

**ATTRIBUTES IN HTML**

```python
tag=soup.div
print(tag.attrs)
```

```python
tag=soup.header
print(tag.attrs)
```

```python
atb=tag.attrs
print(atb["role"])
print(atb["class"])
```

**Navigable Strings in HTML**

```python
tag=soup.div.p
print(tag)
```

```python
tag=soup.div.p
print(tag.string)
```

```python
tag=soup.header.div.a.button.span
print(tag.string)
```

**BeautifulSoup Find() Function**

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup.find("div"))
```

```python
price=soup.find("h4",class_="price float-end card-title pull-right")
print(price)
print(price.string)
```

```python
description=soup.find("p",class_="description card-text")
print(description)
print(description.string)
```

**FindAll() Function of BeautifulSoup**

```python
price=soup.find_all("h4",class_="price float-end card-title pull-right")
print(price)
```

```python
print(price[0].string)
```

```python
print(len(price))
```

```python
for i in price:
  print(i)
  print(i.string)
```

**Find All() Methods**
to push multiple tags

```python
data=soup.find_all(["h4","p","a"])
print(data)
```

```python
data=soup.find_all(string="Galaxy Tab 3")
print(data)
```

```python
import re
data=soup.find_all(string=re.compile("Galaxy"))
print(data)
```
