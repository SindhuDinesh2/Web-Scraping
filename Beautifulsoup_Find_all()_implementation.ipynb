**Beautifulsoup Find_all() implementation with Pandas**

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
```

```python
r=requests.get(url)
print(r)
print(r.text)
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
names=soup.find_all("a",class_="title")
#print(names)
product_name=[]
for i in names:
  name=i.text
  product_name.append(name)
print(product_name)
```

```python
prices=soup.find_all("h4",class_="price float-end card-title pull-right")
#print(prices)
product_price=[]
for i in prices:
  price=i.text
  product_price.append(price)
print(product_price)
```

```python
desc=soup.find_all("p",class_="description card-text")
#print(desc)
product_desc=[]
for i in desc:
  description=i.text
  product_desc.append(description)
print(product_desc)
```

```python
review=soup.find_all("p",class_="review-count float-end")
#print(review)
product_review=[]
for i in review:
  reviews=i.text
  product_review.append(reviews)
print(product_review)
```

```python
df=pd.DataFrame({"Product_name":product_name,"Product_price":product_price,"Product_description":product_desc,"Product_review":product_review})
print(df)
```

```python
df.to_csv("products_dataset.csv")
```
